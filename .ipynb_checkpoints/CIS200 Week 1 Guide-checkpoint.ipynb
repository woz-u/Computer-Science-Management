## 1. What This Course Actually Is (Simplified)

CIS200 is your starting point in understanding what computer science actually is‚Äînot just using technology, but learning how problems are solved through logic, structure, and computation.

In this course, you will begin learning how computer scientists think. That means learning how to:

* Break down a problem into smaller steps

* Describe those steps clearly (this is called designing an algorithm)

* Represent information in a structured way

* Turn your ideas into executable instructions

You are not expected to already know programming or computing theory. What matters is learning the process.

---
### üîç What this course builds toward

By the end of the course, you will be able to:

* Design step-by-step procedures to solve problems

* Use basic procedural programming techniques

* Understand how searching and sorting processes work

* Choose appropriate data types for a given task

* Work confidently with your operating system to save, run, and manage files

These abilities form the foundation that later courses will build on.

---
### üß† Why this actually matters

Every computing-related field‚Äîfrom networking, databases, cybersecurity, AI, and software engineering‚Äîultimately depends on:

* Good problem solving

* Algorithmic thinking

* Representing data properly

* Translating instructions into structured logic

What you learn here is used everywhere else.

You‚Äôre not just completing assignments‚Äîyou are building the beginning of a true CS skillset.

---
### üí¨ What you should expect

You should expect:

* To learn concepts first, then see them applied

* To think before you write

* To revise your approach as you learn better methods

* To get better over time through repetition

You are not expected to be perfect in Week 1.
You are expected to follow directions, try, ask questions correctly, and work consistently.

This is your starting point. If you stay engaged, apply what you learn, and follow the structure provided, CIS200 will not just be a requirement‚Äîit will be the gateway into thinking and working like someone who solves problems using technology.

## 3. How to Stay Ahead (and Not Fall Behind)

One of the hardest parts about this course is that each Exeter LMS Lesson contains multiple Cengage/MindTap units‚Äînot just one. Because of that, it‚Äôs not enough to simply ‚Äúfinish the assignment that is due.‚Äù You need to stay slightly ahead so that the next lesson never begins at zero.

Instead of doing only exactly what is due, a better plan is to **finish one full Cengage unit each day, and also begin the next one**, even if you only complete a small portion of it.

Here is what that looks like in practice during Week 1:

| Day | Recommended Progress |
|------|----------------------|
| Monday (course opens) | Complete Unit 1 + begin a few pages of Unit 2 |
| Tuesday | Finish Unit 2 + begin deeper into Unit 3 |
| Wednesday (Lesson 1 due) | Complete Unit 3 + begin Unit 4 |
| Thursday | Finish most of Unit 4 + begin Unit 5 |
| Friday | Complete Unit 5 + begin Unit 6 |
| Saturday | Finish Unit 6 + begin Unit 7 |
| Sunday | Complete Unit 7 + begin Unit 8 |

This pattern matters.

You are not just finishing a task‚Äîyou are doing **something today that makes tomorrow easier**.

---

### üí° Why this works

If you complete Unit 1 on Monday and start Unit 2, then Tuesday does not begin at zero.  
You simply finish Unit 2 and move into Unit 3.

On Wednesday, you finish Unit 3 and you‚Äôre already partway into Unit 4, meaning that Thursday starts easier.

Every day benefits from yesterday‚Äôs small investment.

This reduces stress dramatically when you hit the midpoint of the course.

---

### üß† The students who fall behind typically:

- Finish the bare minimum only on the due-date day  
- Treat deadlines as starting points rather than finish lines  
- Wait until Wednesday and Sunday evenings  

By that point, the number of units still left becomes overwhelming.

---

### üöÄ The students who succeed usually:
- Do a small amount of work every day  
- Start the next unit right after finishing the current one  
- Enter deadline days already ahead  
- Experience lower stress  
- Maintain consistent energy  

This course does not reward last-minute bursts of effort.  
It rewards **steady, daily forward movement**.

If you begin each day slightly ahead of where you need to be, you will never be surprised, rushed, or overwhelmed at the end of the week.

## 4. Course Best Practices

These best practices will help you avoid common mistakes, stay organized, and work efficiently. Students who consistently follow these guidelines tend to finish the course with less stress and stronger understanding.

### ‚úî Best Practice #1 ‚Äî Name your files clearly and consistently  
Always name files in the format:

                        Lastname_AssignmentName_QuestionNumber.extension


Examples:

                        Hardeman_Unit1Exercises_Q1.docx
                        Hardeman_LabExperience2_Q3.pdf


This helps your files stay organized and prevents confusion later in the term.

---

### ‚úî Best Practice #2 ‚Äî Work locally before submitting  
Always complete your work in a local tool **before** you upload to the LMS.

Working directly inside a browser window or LMS can lead to:
- lost work if the browser crashes  
- partial submissions  
- timing-out  
- files not properly saving  

Working locally first lets you:
- save versions  
- back up your work  
- review before submitting  

---

### ‚úî Best Practice #3 ‚Äî Ask questions with full context  
When seeking help, always include:

- assignment name  
- Cengage unit  
- question number  
- screenshot or file attachment  

This reduces delays and makes it easier for education staff to help you.

Bad request:
> "I don't get this."

Better:
> "I'm on Cengage Unit 4, Question 7 ‚Äî my output is different. Screenshot included."

---

### ‚úî Best Practice #4 ‚Äî Know your deadlines ahead of time  

Deadlines are always:
- **Wednesday**
- **Sunday**

Write these down.
Plan ahead.
Do not start assignments on the day they are due.

---

### ‚úî Best Practice #5 ‚Äî Use external resources responsibly  

This course is **open-resource friendly**:
- You may use AI
- You may search online
- You may read documentation
- You may research examples

This is a **blessing** when you deepen your understanding.  
It becomes a **curse** when you copy answers without learning the steps behind them.

Computer scientists constantly leverage the internet; the difference is that they understand what they find‚Äînot just paste solutions.

---

### ‚úî Best Practice #6 ‚Äî Keep moving forward when stuck  
If you are stuck too long, move on.

Your priority in this course is simple:
### üéØ Reach at least a 70% passing score.

Then:
- strengthen weak areas later  
- revisit lessons after passing  
- reinforce understanding at your pace  

Do not lose momentum over a single question.  
Progress > Perfection.

---

Students who consistently follow these best practices:
- stay ahead  
- reduce stress  
- avoid last-minute scrambles  
- complete the course with confidence
